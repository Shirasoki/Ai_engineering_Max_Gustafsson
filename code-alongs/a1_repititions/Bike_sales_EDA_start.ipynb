```python
import pandas as pd

df = pd.read_csv("../data/Sales.csv", index_col=0, parse_dates=True)
df.head()
```

```python
yearly_sales = df.groupby("Year")["Profit"].sum()
print(yearly_sales)
```

```python
country_sales = df.groupby("Country")["Profit"].sum().sort_values(ascending=False)
print(country_sales)
```
